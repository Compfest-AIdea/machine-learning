```python
import tensorflow as tf
from tensorflow.keras.saving import load_model

model = load_model("/content/machine-learning/Models/scalp_model.h5")
```

## Alopecia

```python
import shutil
import os

test_dir = '/content/machine-learning/Dataset/Dataset-Image-Augmented/test/alopecia-areata'
alopecia_path = []

for path in os.listdir(test_dir):
  full_path = os.path.join(test_dir,path)
  alopecia_path.append(full_path)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

alopecia_confidence = []
i = 1
#['normal', 'alopecia-areata', 'tinea-capitis', 'scalp-psoriasis', 'seborrhoeic-dermatitis']
disease_mapper = {
    0: 'alopecia-areata',
    1: 'normal',
    2:'scalp-psoriasis',
    3:'seborrhoeic-dermatitis',
    4:'tinea-capitis'}
for path in alopecia_path:
    img = load_img(path, target_size=(150, 150))
    img_array = img_to_array(img)
    img_array /= 255.0
    img_array = np.expand_dims(img_array, axis=0)

    y_pred_single = model.predict(img_array)
    y_pred_single_class = np.argmax(y_pred_single, axis=1)
    y_pred_single_value = np.max(y_pred_single)

    alopecia_confidence.append(
        disease_mapper[y_pred_single_class[0]]
    )

    i += 1
```

```python
from collections import Counter

# Menghitung kemunculan tiap elemen unik
count = Counter(alopecia_confidence)

# Menampilkan hasilnya
print(f"Kelas aktual: alopecia-areata \n")
print(f"Kelas prediksi:")
for elemen, kemunculan in count.items():
    print(f"{elemen}: {kemunculan}")
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd

# Daftar kelas aktual (ground truth)
alopecia_aktual = ['alopecia-areata'] * len(alopecia_confidence)

# Daftar kelas prediksi
alopecia_prediksi = alopecia_confidence
```

```python
kelas_aktual = alopecia_aktual + normal_aktual + psoriasis_aktual + seboroik_aktual + tinea_aktual
kelas_prediksi = alopecia_confidence + normal_confidence + psoriasis_confidence + seboroik_confidence + tinea_confidence

kelas_unik = np.unique(kelas_prediksi)

# Menghitung confusion matrix
confusion_mtx = confusion_matrix(kelas_aktual, kelas_prediksi, labels=kelas_unik)
```

```python
# Fungsi untuk menggambar confusion matrix
import matplotlib.pyplot as plt
import itertools

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
# Menampilkan confusion matrix
plt.figure()
plot_confusion_matrix(confusion_mtx, kelas_unik, normalize=False, title='Confusion Matrix')
plt.show()
```

```python
# Menampilkan confusion matrix
plt.figure()
plot_confusion_matrix(confusion_mtx, kelas_unik, normalize=True, title='Confusion Matrix')
plt.show()
```
